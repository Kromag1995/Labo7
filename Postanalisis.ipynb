```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
f = open('./post_analisis/Complejidades_26.txt')
Complejidades = f.read()
```

```python
comp = {}
for linea in Complejidades.split('\n')[1:-1]:
    l = linea.split(";")
    comp[l[0]] = {"media": l[2], "mediana": l[3]}
```

```python
llaves_ordenadas = sorted(comp)
llaves_v = llaves_ordenadas[::2]
llaves_t = llaves_ordenadas[1::2]
```

```python
llaves_t[60]
```

```python
frecuencia = [250,230,210,190,170, 150]
```

```python
for i in range(60):
    plt.plot(frecuencia[i//10],float(comp[llaves_t[i]]["media"]), "r.")
    plt.plot(frecuencia[i//10],float(comp[llaves_t[i]]["mediana"]), "b.")
plt.plot(frecuencia[i//10],float(comp[llaves_t[i]]["media"]), "r.", label="Media")
plt.plot(frecuencia[i//10],float(comp[llaves_t[i]]["mediana"]), "b.", label="Mediana")
plt.plot(frecuencia, 0.5*np.ones(len(frecuencia)), 'k-.')
plt.legend(loc='best')
plt.grid()
plt.title("Series t")
plt.xticks(frecuencia,frecuencia[:-1]+["170 ancho 2.5"])
plt.ylabel("complejidad(ua)")
plt.xlabel("frecuencia(kHz)")
#plt.savefig("series_t.png")
```

```python
for i in range(60):
    plt.plot(frecuencia[i//10],float(comp[llaves_v[i]]["media"]), "r.")
    plt.plot(frecuencia[i//10],float(comp[llaves_v[i]]["mediana"]), "b.")
plt.plot(frecuencia[i//10],float(comp[llaves_v[i]]["media"]), "r.", label="Media")
plt.plot(frecuencia[i//10],float(comp[llaves_v[i]]["mediana"]), "b.", label="Mediana")
plt.plot(frecuencia, 0.5*np.ones(len(frecuencia)), 'k-.')
plt.legend(loc='best')
plt.title("Series V")
plt.grid()
plt.xticks(frecuencia,frecuencia[:-1]+["170 ancho 2.5"])
plt.ylabel("complejidad(ua)")
plt.xlabel("frecuencia(kHz)")
plt.savefig("series_v.png")
plt.show()
```

```python
f = open('./post_analisis/resultados_2.csv')
resultados_nists = f.read()
```

```python
sr = resultados_nists.split('\n')[1:-2]
```

```python
comp = {}
for linea in sr:
    l = linea.split(";")
    nombre = l[0][:18].replace('20210407-','')
    comp[nombre] = []
    for i in range(1,len(l)):
        resultado = l[i].split(',')[-1].replace(')','').replace(']','').strip()
        comp[nombre].append(resultado)
```

```python
llaves_ordenadas = sorted(comp)
llaves_v = llaves_ordenadas[::2]
llaves_t = llaves_ordenadas[1::2]
```

```python
for i in llaves_t:
    #print(comp[i])
    count = 0
    for element in comp[i]:
        if element == 'True':
            count += 1
    if count > 5:
        print(count)
        print(i)
```

```python
for i in llaves_v:
    #print(comp[i])
    count = 0
    for element in comp[i][:-2]:
        if element == 'True':
            count += 1
    if count > 4:
        print(count)
        print(i)
```
